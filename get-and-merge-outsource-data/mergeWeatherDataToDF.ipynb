```python
import numpy as np
import pandas as pd
```

# Read Cleaned Data

```python
df = pd.read_csv('./../clean-data/df_clean.csv')

df.head(3)
```

# Read all_district_weather_data

```python
weather_data = pd.read_csv('./all_district_weather_data.csv')
```

### convert Date_Time into timestamp

```python
weather_data['timestamp'] = pd.to_datetime(weather_data['Date_Time'], format='mixed', utc=True)

weather_data.head(3)
```

```python
# drop Date_Time column

weather_data['timestamp'] = pd.to_datetime(weather_data['Date_Time'], format='mixed', utc=True)

weather_data = weather_data.drop("Date_Time", axis=1)

weather_data.head(3)
```

# Merge df and weather_data

### using same district and timestamp

```python
# Create date column
df['date'] = pd.to_datetime(df['timestamp'], format='mixed', utc=True).dt.date
weather_data['date'] = pd.to_datetime(weather_data['timestamp'], format='mixed', utc=True).dt.date

# Merge
merged = df.merge(
    weather_data[['district','date','PM2.5','WS','WD','Temp','RH','BP']],
    on=['district','date'],
    how='left'
)

# Drop 'date' column
merged = merged.drop(columns=['date'])
```

```python
# Check result

merged.head()
```

```python
merged.info()
```

### Null count

```python
print('Count null PM2.5:', merged['PM2.5'].isna().sum())
print('Count null WS:   ', merged['PM2.5'].isna().sum())
print('Count null WD:   ', merged['PM2.5'].isna().sum())
print('Count null Temp: ', merged['PM2.5'].isna().sum())
print('Count null Temp: ', merged['PM2.5'].isna().sum())
print('Count null RH:   ', merged['PM2.5'].isna().sum())
print('Count null BP:   ', merged['PM2.5'].isna().sum())
```

# Drop Row where weather data is null

```python
print(merged['district'].nunique())
print(merged['district'].value_counts().head(10))
```

```python
weather_cols = ['PM2.5', 'WS', 'WD', 'Temp', 'RH', 'BP']

# change dtype from object -> float before drop null
for col in weather_cols:
    merged[col] = pd.to_numeric(merged[col], errors='coerce')

df_clean = merged.dropna(subset=weather_cols, how='any')

df_clean.info()
```
