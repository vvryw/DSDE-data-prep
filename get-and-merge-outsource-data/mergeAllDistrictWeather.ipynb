```python
import os
import pandas as pd
from tqdm import tqdm
```

# Load mapping file

```python
df_map = pd.read_csv("district_mapping_clean.csv", dtype={"id": str})
df_map["name"] = df_map["name"].astype(str).str.strip()


# Folder containing weather CSVs
DATA_DIR = "district-weather-data"

# List to collect all districts' DataFrames
all_dfs = []
```

```python
# Loop through each district name

for _, row in tqdm(df_map.iterrows(), total=df_map.shape[0]):
    district_name = row["name"]

    filename = f"{district_name}_weather_data.csv"
    filepath = os.path.join(DATA_DIR, filename)

    if not os.path.exists(filepath):
        print(f"Missing file: {filepath}")
        continue

    try:
        df = pd.read_csv(filepath)
        df["district"] = district_name  # ensure correct district name
        all_dfs.append(df)
    except Exception as e:
        print(f"Error reading {filepath}: {e}")
```

# Combine all districts

```python
if len(all_dfs) == 0:
    print("No files loaded. Check mapping or folder path.")
else:
    merged_df = pd.concat(all_dfs, ignore_index=True)

    # Save final combined CSV
    merged_df.to_csv("all_district_weather_data.csv", index=False, encoding="utf-8-sig")

    print("Combined file saved as: all_district_weather_data.csv")
    print("Total rows:", merged_df.shape[0])
```
