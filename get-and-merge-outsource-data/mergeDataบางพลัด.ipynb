```python
import pandas as pd
import os
```

```python
df_bp = pd.read_csv('./district-weather-data/บางพลัด0_weather_data.csv')

df_bp2 = pd.read_csv('./district-weather-data/สวนหลวงพระราม_8_บางพลัด_weather_data.csv')
```

```python
df_bp.head(5)
```

```python
df_bp2.head(5)
```

Merge DataFrame

df_bp  => district	Date_Time	PM2.5	WS

df_bp2 => WD	    Temp	    RH	    BP

```python
# [Drop WD, Temp, RH, BP] from df_bp
drop_cols = ['WD', 'Temp', 'RH', 'BP']
df_bp_clean = df_bp.drop(columns=[c for c in drop_cols if c in df_bp.columns])

df_bp_clean.head(5)
```

```python
# Merge with df_bp2 (use only WD, Temp, RH, BP)
merged_df_bp = df_bp_clean.merge(
    df_bp2[['Date_Time', 'WD', 'Temp', 'RH', 'BP']],
    on='Date_Time',
    how='left'
)
```

```python
merged_df_bp.head(5)
```

Save into CSV

```python
district_name = merged_df_bp['district'].iloc[0]
filename = f"{district_name}_weather_data.csv"
filepath = os.path.join("district-weather-data", filename)

merged_df_bp.to_csv(filepath, index=False, encoding="utf-8-sig")

print(f"Saved: {filepath}")
```
