```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('bangkok_traffy.csv')

df.head(3)
```

```python
print("shape:", df.shape)
df.info()

# coords, timestamp, ticket_id
```

Delete unused column

```python
# drop unused column
# star, photo, photo_after

drop_col = ['photo', 'photo_after', 'star', 'ticket_id']

df = df.drop(columns=drop_col, errors='ignore')

print(df.shape)
print(df.columns)
```

Convert timestamp and last_activity column

```python
# convert timestamp to yyyy-mm-dd hh-mm-ss

df['timestamp'] = pd.to_datetime(df['timestamp'], format='mixed', utc=True)

print(df['timestamp'].head(3))
# print('====================')
df.head(3)
```

```python
# delete data before 2023-01-01
# delete data after 2024-12-31

print('before:', df.shape)

df = df[(df['timestamp'] >= '2023-01-01') & (df['timestamp'] < '2025-01-01')]

print('after:', df.shape)
print(df['timestamp'].value_counts())
df.head(3)
```

```python
# convert last_activity to yyyy-mm-dd hh-mm-ss

df['last_activity'] = pd.to_datetime(df['last_activity'], format='mixed', utc=True)

print(df['last_activity'].head(3))
# print('====================')
df.head(3)
```

Check total empty tyoe and comment row

```python
# count empty type

empty_type = df[
    df['type'].isna() | 
    (df['type'].str.strip() == "") | 
    (df['type'].str.strip() == "{}")
]

total_row = df.shape[0]
total_empty_type = empty_type.shape[0]

print('total row:', total_row)
print('total empty:', total_empty_type)
print('percentage:', round(total_empty_type / total_row, 5))
empty_type.head(3)
```

```python
# count empty comment

empty_comment = df[
    df['comment'].isna() | 
    (df['comment'].str.strip() == "") | 
    (df['comment'].str.strip() == "{}")
]

total_row = df.shape[0]
total_empty_comment = empty_comment.shape[0]

print('total row:', total_row)
print('total empty:', total_empty_comment)
print('percentage:', round(total_empty_comment / total_row, 5))
empty_comment.head(3)
```

Drop empty type and comment

```python
print('before:', df.shape)

ts_empty_comment = empty_comment['timestamp']
ts_empty_type = empty_type['timestamp']

# combine todrop row
ts_to_drop = pd.concat([ts_empty_comment, ts_empty_type]).unique()

# drop rows where timestamp matches any empty timestamp
df = df[~df['timestamp'].isin(ts_to_drop)]

print('after:', df.shape)

print(df.info())
```

Get only row with state is 'เสร็จสิ้น'

```python
df = df[df['state'] == 'เสร็จสิ้น']

df.shape
```

Add 'duration' column
last_activity - timestamp

```python
df['duration'] = df['last_activity'] - df['timestamp']

df.head(3)
```
