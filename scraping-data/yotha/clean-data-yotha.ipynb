# Import Lib

```python
import pandas as pd
import os
```

# Read csv

```python
df = pd.read_csv('data_bangkok_yota_road_2025.csv')
```

```python
print(df.shape)
print(df.head(3))
```

# Drop unused columns

```python
drop_cols = ['ROAD_NO' , 'SECTION_START' , 'SECTION_END' ,
             'WIDTH' , 'ANNOUNCE_DATE' , 'DEPARTMENT' , 'ROAD_OWNER' ]

df = df.drop(columns=drop_cols, axis=1)
```

```python
df.head(3)
```

```python
df['DISTRICT'].unique()
```

# Separate District

```python
df['DISTRICT'].nunique()
```

```python
df['DISTRICT'] = df['DISTRICT'].str.split(',')
df = df.explode('DISTRICT')
df['DISTRICT'] = df['DISTRICT'].str.strip()

print(df['DISTRICT'].nunique())
df['DISTRICT'].unique()
```

# Get Road

```python
df_road = df[df['ROAD_TYPE'] == 'ถนน']

df_road = df_road.drop(columns='ROAD_TYPE', axis=1)
df_road.shape
```

## Groupby Road_Name & District

```python
df_road.head(5)
```

```python
df_road_sum = (
    df_road.groupby(['ROAD_NAME', 'DISTRICT'], as_index=False)
      .agg({'LENGTH': 'sum'})
)

print(df_road_sum.shape)
df_road_sum.head()
```

# Get ซอย

```python
df_alley = df[df['ROAD_TYPE'] == 'ซอย']

df_alley = df_alley.drop(columns='ROAD_TYPE', axis=1)


print(df_alley.shape)
print(df_alley.head())
```

## Groupby Road_Name & District

```python
df_alley_sum = (
    df_alley.groupby(['ROAD_NAME', 'DISTRICT'], as_index=False)
      .agg({'LENGTH': 'sum'})
)

print(df_alley_sum.shape)
df_alley_sum.head()
```

# Merge df_road_sum & df_alley_sum

```python
total_length_each_road_alley = pd.concat([df_alley_sum, df_road_sum], ignore_index=True)
```

```python
total_length_each_road_alley.shape
```

# Save csv

## Save csv total_length_each_road_alley

```python
os.makedirs("cleaned-data", exist_ok=True)

total_length_each_road_alley.to_csv("cleaned-data/total_length_each_road_alley.csv", index=False, encoding='utf-8-sig')
```

## Save csv df_road_sum

```python
df_road_sum.to_csv("cleaned-data/df_road_sum.csv", index=False, encoding='utf-8-sig')
```

## Save csv df_alley_sum

```python
df_alley_sum.to_csv("cleaned-data/df_alley_sum.csv", index=False, encoding='utf-8-sig')
```
